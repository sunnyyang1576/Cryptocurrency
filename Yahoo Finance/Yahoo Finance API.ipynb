### Get all crypto ticker

```python
with open("data/yahoo_finance_crypto_list.txt") as f:
    
    ticker_list = f.readlines()
    
    ticker_list = [ticker.strip() for ticker in ticker_list]
```

### Get All Crypto Price

```python
start_date = "01/01/2015"
end_date = "01/01/2020"
```

```python
price_dic = {}

for ticker in ticker_list:
    
    price_df = get_data(ticker=ticker,
                        start_date=None,
                        end_date=None,
                        index_as_date=True,
                        interval="1d")
    
    price_dic[ticker] = price_df
    
    time.sleep(1)
    
    
```

```python
price_dic["BTC-USD"]
```

```python
price_df_list = [price_dic[ticker] for ticker in ticker_list]
```

```python
price_df = pd.concat(price_df_list,axis=0)
```

```python
price_df.to_csv("data/yahoo_finance_crypto_price_df.csv")
```
