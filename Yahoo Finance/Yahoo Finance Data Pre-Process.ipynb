```python
import pandas as pd
import numpy as np
```

```python
def wide_table_reframe(df,colname):
    
    long_df = pd.pivot_table(df,values=colname,index="date",columns="ticker")
    
    return long_df


def asset_filter(df,filter_date,bottom_threshold):
    
    signal_series = df.loc[filter_date,:].dropna()
    
    filter_series = signal_series[(signal_series>=bottom_threshold)]
    
    filter_list = list(filter_series.index)
    
    return filter_list
```

```python
data_directory = "data/"
file = "yahoo_finance_crypto_price_df.csv"
```

```python
df = pd.read_csv(data_directory+file)
df["date"] = pd.to_datetime(df["date"])
```

```python
df["date"] = pd.to_datetime(df["date"])
```

```python
ticker_list = list(set(df["ticker"]))
```

```python
volume_df = wide_table_reframe(df,colname="volume")
price_df = wide_table_reframe(df,colname="close")
```

```python
asset_list = asset_filter(price_df,filter_date="2017-12-30",bottom_threshold=0.01)
```

```python
sub_volume_df = volume_df[asset_list][volume_df.index>"2017-12-30"]
```

```python
threshold = sub_volume_df.median().describe()["25%"]
```

```python
asset_list = list(sub_volume_df.median()[sub_volume_df.median()>threshold].index)
```

```python
len(asset_list)
```

```python
sub_price_df = price_df[asset_list][price_df.index>"2017-12-30"]
```

```python
sub_price_df = sub_price_df.fillna(method="ffill")
```

```python
sub_price_df.to_csv("yahoo_data_filter_0_0_1.csv")
```

```python
sub_volume_df = volume_df[asset_list][price_df.index>"2017-12-30"]
```

```python
sub_volume_df = sub_volume_df.fillna(method="ffill")
```
