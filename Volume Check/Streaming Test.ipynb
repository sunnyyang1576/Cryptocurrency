```python
import pandas as pd
import numpy as np

from src.asset_filter import UniverseFilter
from src.signal_creator import SignalCreator_Momentum
from src.strategy_selection import Strategy
from src.price_dataframe import PriceDataFrame
from src.order_creator import OrderCreator
```

```python
class AssetEvaluator():
    
    
    def __init__(self,asset_list,weight_list):
        
        self.asset_list = asset_list
        self.weight_list = pd.Series(weight_list)
        self.weight_list.index = asset_list
        
    def portfolio_return(self,asset_return_series):
        
        ret_series = asset_return_series[self.asset_list]
        
        tot_ret = (ret_series * self.weight_list).sum()
        
        return tot_ret
```

```python
def one_round_strategy(price_df_wide,vol_df_wide,asset_universe):
    
    asset_filter = UniverseFilter(asset_universe)
    asset_filter.vol_filter(vol_df_wide.iloc[-1,:],threshold="25%")
    update_universe = asset_filter.update_universe
    
    signal_creator = SignalCreator_Momentum(update_universe,price_df_wide)
    signal_1 = signal_creator.create_signal(look_back=1)
    signal_3 = signal_creator.create_signal(look_back=3)
    
    strategy_selector = Strategy()
    
    top_port_2,bottom_port_2 = strategy_selector.topn_double_signal_selection(signal_1,signal_3,n_1=20,n_2=10)
    
    top_order = OrderCreator(top_port_2)
    top_order.equal_weight()
    top_order = top_order.create_order_dictionary()
    
    bottom_order = OrderCreator(bottom_port_2)
    bottom_order.equal_weight()
    bottom_order = bottom_order.create_order_dictionary()
    
    return (top_order,bottom_order)


def evaluate(top_order,bottom_order,ret_series):
    
    evaluator = AssetEvaluator(top_order["asset_list"],top_order["weight"])
    top_ret = evaluator.portfolio_return(ret_series)

    evaluator = AssetEvaluator(bottom_order["asset_list"],bottom_order["weight"])
    bottom_ret = evaluator.portfolio_return(ret_series)
    
    return (top_ret,bottom_ret)
```

```python
df = pd.read_csv("data/yahoo_finance_data.csv")
df["date"] = pd.to_datetime(df["date"])

vol_df = df.pivot_table(columns="ticker",values="volume",index="date")
price_df = df.pivot_table(columns="ticker",values="price",index="date")
```

```python
def streaming_data_generator(price_df,vol_df,length=5):
    
    ret_df = price_df.pct_change()
    
    for idx in range(length+1,price_df.shape[0]):
        
        sub_price_df = price_df[(idx-length):(idx)]
        sub_vol_df = vol_df[(idx-5):(idx)]
        ret_series = ret_df.iloc[idx,:]
        
        yield (sub_price_df,sub_vol_df,ret_series)
        
        
```

```python
data_generator = streaming_data_generator(price_df,vol_df,length=5)
```

```python
asset_universe = list(price_df.columns)

top_ret_list = []
bottom_ret_list = []
date_list = []

for price_sub,vol_sub,ret_series_sub in data_generator:
    
    date = ret_series_sub.name
    
    print(date)
    
    top_order,bottom_order = one_round_strategy(price_sub,vol_sub,asset_universe)
    
    top_ret,bottom_ret = evaluate(top_order,bottom_order,ret_series_sub)
    
    top_ret_list.append(top_ret)
    bottom_ret_list.append(bottom_ret)
    date_list.append(date)
```

```python
top_ret = pd.Series(top_ret_list)
bottom_ret = pd.Series(bottom_ret_list)
date_list = pd.Series(date_list)
```

```python
df = pd.concat([date_list,top_ret,bottom_ret],axis=1)
df.columns = ["date","top","bottom"]
df = df.set_index("date")
```
